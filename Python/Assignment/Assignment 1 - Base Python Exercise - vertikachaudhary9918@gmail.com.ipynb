```python
def fn_describe(args):
    sum_input = sum(args)
    count_input = len(args)
    min_input = np.min(args)
    max_input = np.max(args)
    mean_input = sum_input/count_input
    
     
    return [sum_input,count_input,min_input,max_input,mean_input]
    
```

```python
fn_describe([2,3,4,5])
```

```python
def area_of_shape(shape,**kwargs):
    area = 0
    if shape == 'square':
        area = kwargs['side']*kwargs['side']
    elif shape == 'circle':
        area = 3.14 * kwargs['radius']**2
    elif shape == 'rectangle':
        area = kwargs['length'] * kwargs['breath']
    return area
```

```python
area_of_shape('square',side=5)
```

```python
area_of_shape('circle',radius=25)
```

```python
area_of_shape('rectangle',length=5,breath = 10)
```

```python
Q12: Write a UDF to reconcile the values within two lists.
    l1 = ['January', 'February', 'March', 'May', 'June', 'September', 'December']
    l2 = ['January', 'February', 'April', 'June', 'October', 'December']

Hint:
-----
def func(l1, l2):
    your code here...
    
Output:
{'Matched': ['January', 'February', 'June', 'December'],
    'Only in l1': ['March', 'May', 'September'],
        'Only in l2': ['April', 'October']}
```

```python
l1 = ['January', 'February', 'March', 'May', 'June', 'September', 'December']
l2 = ['January', 'February', 'April', 'June', 'October', 'December']

def reconcile(s1,s2):
    set1 = set(s1)
    set2 = set(s2)
    
    Match = set1 & set2
    print("Match:",Match)
    only_set1 = set(s1) - set(s2)
    print("Only in l1:",only_set1)
    only_set2 = set(s2) - set(s1)
    print("Only in l2:",only_set2)
```

```python
reconcile(l1,l2)
```

```python
def test_prime(n):
    if (n==1):
        x = "not prime"
    elif (n==2):
        x ="prime"
    else:
        for x in range(2,n):
            if(n % x==0):
                x = "prime"
            else:
                x = "not prime"
    return x
print(test_prime(11))
    
```

```python
# Python program to find the factorial of a number provided by the user.

# change the value for a different result
num = 7

# To take input from the user
#num = int(input("Enter a number: "))

factorial = 1

# check if the number is negative, positive or zero
if num < 0:
    print("Sorry, factorial does not exist for negative numbers")
elif num == 0:
    print("The factorial of 0 is 1")
else:
    for i in range(1,num + 1):
        factorial = factorial*i
    print("The factorial of",num,"is",factorial)
```

```python
def string_test(s):
    d={"UPPER_CASE":0, "LOWER_CASE":0}
    for c in s:
        if c.isupper():
            d["UPPER_CASE"]+=1
        elif c.islower():
            d["LOWER_CASE"]+=1
        else:
            pass
    print ("Original String : ", s)
    print ("No. of Upper case characters : ", d["UPPER_CASE"])
    print ("No. of Lower case Characters : ", d["LOWER_CASE"])

string_test('Indian is my Nationality')
```
